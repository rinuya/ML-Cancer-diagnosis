# Data exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

%matplotlib inline
```

```python
data = load_breast_cancer()

X = data.data
y = data.target

feature_names = data.feature_names
target_names = data.target_names

# 212 samples are Malignant, 357 are benging Benign 
# -> 0 == malignant, 1 == benign

print(f"Number of samples: {X.shape[0]}")
print(f"Number of features: {X.shape[1]}")
print(f"\nFeature names:\n- {"\n- ".join(feature_names)}")
print(f"\nTarget names:\n- {"\n- ".join(target_names)}")
```

```python
# Let's check what the data looks like
X[0], X[0][0], y[0]
```

```python
df = pd.DataFrame(X, columns=feature_names)
df["target"] = y
df.head()
```

```python
df.describe()
```

```python
# seperate histogram per feature
features = df.columns[:-1]  # Exclude the target column
num_features = len(features)

plt.figure(figsize=(15, 30))
for i, feature in enumerate(features):
    plt.subplot((num_features // 3) + 1, 3, i + 1)
    for label in [0, 1]:
        subset = df[df['target'] == label]
        plt.hist(subset[feature], bins=20, alpha=0.5, label=f'{"malignant" if label == 0 else "benign"}')
    plt.title(feature)
    plt.legend()
    
plt.tight_layout()
plt.show()
```
